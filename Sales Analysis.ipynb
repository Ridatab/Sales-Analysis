# Analysis 

```python
# 15. Best-Selling Product
sales.groupby('Product')['Quantity Ordered'].count().plot(kind = 'bar')
```

```python
# 16. Sales for Each Product
sales.groupby('Product')['TotalSales'].sum().plot(kind = 'bar')
```

```python
# 17. Average Order Value
sales.groupby('Product')['Quantity Ordered'].mean().plot(kind ='barh')
```

```python
# 18. Sales Trends
Sale_Trend = sales.pivot_table( columns = ['Month'], values = 'Price Each', aggfunc = 'sum')
SaleTrend = Sale_Trend.T
SaleTrend
```

```python
x = sales['Month']
y = sales['Price Each']

plt.title('Sales Trend')
plt.scatter(x,y)
plt.show()
```

```python
# 19. Monthly sales 
sales.groupby('Month')['TotalSales'].sum().plot(kind = 'line')
```

```python
# Monthly Orders
x = sales['Month']
y = sales['Order ID']

plt.title("Monthly Orders")
plt.bar(x,y)
plt.show()
```

```python
# 21. Citywise Sales
df = sales.pivot_table(columns = 'City', values = 'TotalSales', aggfunc = 'sum')
City_Sales = df.T
City_Sales
```

```python
# 22. City wise Orders
df = sales.pivot_table(columns = 'City', values = 'Order ID', aggfunc = 'count')
City_Sales = df.T
City_Sales
```
